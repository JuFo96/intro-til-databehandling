```python
from intro_pandas import get_path, plot_home_types, plot_regional_prices
import pandas as pd
import matplotlib.pyplot as plt

input_file = get_path("../Data/DKHousingPricesSample100k.csv")
```

This is generally a duplicate of the intro_pandas.py file, just to showcase some of the results in a single file rather than relying on terminal output and files being written.

```python
df = pd.read_csv(input_file)
df.head(10)
```

```python
regional_prices = df.groupby("region")["purchase_price"].mean()
print(regional_prices)
```

```python
plt.figure()
plt.bar(regional_prices.index, regional_prices.values)
plt.title("Average Purchase Price by Region")
plt.ylabel("Average Price [DKK]")
plt.tight_layout()
```

```python
home_types = df["house_type"].value_counts()

plt.figure()
home_types.plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribution of House Types")
plt.ylabel("")
plt.tight_layout()
```
